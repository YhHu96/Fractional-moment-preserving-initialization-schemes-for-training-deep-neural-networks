```python
import numpy as np
from scipy.special import gamma, comb
```

```python
# s=1 case
data=[]
s=1
for d in range(30):
    temp=0
    for n in range(d+1):
        temp=temp+comb(d+1,n+1)*1/(2**(d+1))*gamma((n+1)/2+s/2)/gamma((n+1)/2)
    temp=temp*pow(2,s/2)
    sigma=1/(pow(temp,1/s))
    data.append(np.log(sigma))

# s=1.5 case
data15=[]
s=1.5
for d in range(30):
    temp=0
    for n in range(d+1):
        temp=temp+comb(d+1,n+1)*1/(2**(d+1))*gamma((n+1)/2+s/2)/gamma((n+1)/2)
    temp=temp*pow(2,s/2)
    sigma=1/(pow(temp,1/s))
    data15.append(np.log(sigma))
    
# s=2 case
data2=[]
s=0.1
for d in range(30):
    temp=0
    for n in range(d+1):
        temp=temp+comb(d+1,n+1)*1/(2**(d+1))*gamma((n+1)/2+s/2)/gamma((n+1)/2)
    temp=temp*pow(2,s/2)
    sigma=1/(pow(temp,1/s))
    data2.append(np.log(sigma))
    
# s=0.5 case
data0=[]
s=0.5
for d in range(30):
    temp=0
    for n in range(d+1):
        temp=temp+comb(d+1,n+1)*1/(2**(d+1))*gamma((n+1)/2+s/2)/gamma((n+1)/2)
    temp=temp*pow(2,s/2)
    sigma=1/(pow(temp,1/s))
    data0.append(np.log(sigma))
```

```python
# log sigma figure vs d
import matplotlib.pyplot as plt
figsize = 8,6
figure, ax = plt.subplots(figsize=figsize)
plt.plot(data0,linewidth=3.5)
plt.plot(data,linewidth=3.5)
plt.plot(data15,linewidth=3.5)
plt.plot(data2,linewidth=3.5)
# plt.title('s=1',fontsize=20)
plt.ylabel(r'$log(\sigma)$',fontsize=30)
plt.xlabel('d',fontsize=30)
plt.tick_params(labelsize=30)
plt.legend(['s=0.5','s=1','s=1.5','s=2'], loc='upper right',fontsize=30)
labels = ax.get_xticklabels() + ax.get_yticklabels()
[label.set_fontname('Times New Roman') for label in labels]
plt.show()
```
